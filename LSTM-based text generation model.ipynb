```python
import nltk
import numpy as np
import tensorflow as tf
from nltk.tokenize import word_tokenize
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.optimizers import Adam

# 1. Download NLTK resources
nltk.download('punkt')

# 2. Input text
text = """
Deep learning models generate meaningful text using LSTM networks.
LSTM models are effective for sequence prediction tasks in NLP.
"""

# 3. Preprocessing using NLTK
tokens = word_tokenize(text.lower())

# Build vocabulary
vocab = sorted(set(tokens))
word_to_idx = {word: idx for idx, word in enumerate(vocab)}
idx_to_word = {idx: word for word, idx in word_to_idx.items()}

vocab_size = len(vocab)

# 4. Create input-output sequences
seq_length = 4
X = []
y = []

for i in range(len(tokens) - seq_length):
    X.append([word_to_idx[w] for w in tokens[i:i + seq_length]])
    y.append(word_to_idx[tokens[i + seq_length]])

X = np.array(X)
y = np.array(y)

# 5. Build LSTM model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=32, input_length=seq_length),
    LSTM(128),
    Dense(vocab_size, activation="softmax")
])

model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss="sparse_categorical_crossentropy"
)

model.summary()

# 6. Train the model
model.fit(X, y, epochs=100, batch_size=8)

# 7. Text generation function

def generate_text(model, seed_text, length=20):
    result = word_tokenize(seed_text.lower())

    for _ in range(length):
        seq = result[-seq_length:]
        seq = np.array([[word_to_idx[w] for w in seq]])

        predictions = model.predict(seq, verbose=0)
        next_word_index = np.argmax(predictions)

        result.append(idx_to_word[next_word_index])

    return " ".join(result)

# 8. Generate text
# Use a seed text that exists in the vocabulary
seed_text = "deep learning models generate"
generated_output = generate_text(model, seed_text, length=25)

print("\nGenerated Text:")
print(generated_output)
```
